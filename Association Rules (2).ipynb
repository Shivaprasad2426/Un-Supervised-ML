```python
APPLICATIONS OF APRIORI,ASS
```

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
rules=rules.sort_values('lift',ascending=False)
rules.head()
```

1.Lift is the ratio of the observed support to that expected if A and B were independent. It indicates the strength of a rule over random co-occurrence.

 Lift(A → B) = Support(A ∪ B) / (Support(A) × Support(B))

2 .Support
Support is the proportion of transactions in the database in which the itemset appears. It indicates how frequently an itemset appears in the dataset.

Support(A) = Number of transactions containing A / Total number of transactions

Confidence
Confidence is the likelihood that a rule is correct for a new transaction containing the antecedent.

Confidence(A → B) = Support(A ∪ B) / Support(A)
